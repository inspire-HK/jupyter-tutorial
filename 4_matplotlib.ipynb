```python
### Matplotlib is the common plotting library
```

```python
import numpy as np

t = np.linspace(-100, 100, 200)
func1 = (t ** 2) * np.sin(t) # define f1(t) = t^2 * sin(t)
func2 = (t ** 2) # f2(t) = t^2
```

```python
from matplotlib import pyplot as plt 

plt.figure()
plt.plot(t, func1, 'r-', linewidth=2) # plot func1 in solid red line
plt.plot(t, func2, 'k--', linewidth=2) # plot func2 in dashed black line
plt.xlabel('time') # add x-axis label
plt.ylabel('functions') # add y-axis label
plt.legend(['f1(t)', 'f2(t)']) # add legends
plt.grid(True) # turn on grid lines
plt.show() # show the plot (for non-jupyter, jupyter automatically shows the plot)
```
